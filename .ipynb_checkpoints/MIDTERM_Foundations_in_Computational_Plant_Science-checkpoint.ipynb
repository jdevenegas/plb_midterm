# Datasets


The data you will be analyzing represents leaf shapes measured from a germplasm repository of the USDA, a vineyard in Geneva, NY with accessions of North American grapevine species (genus *Vitis*). Wine grapes come from *Vitis vinifera*, a Eurasian species, but the North American grapevines still contribute to wine production. Nearly all vines used for wine and table grape production are grafted to North American species for disease resistance.

Grapevines are perennial, meaning they continue to grow from year-to-year. Because the same vine remains year after year, they can be sampled and evaluated with respect to changing climate conditions. Plants and agriculture will be suceptible to climate change, but perennial species are uniquely poised to show the how plants respond to a changing climate using the same plants.

In the publication ["Climate and Developmental Plasticity: Interannual Variability in Grapevine Leaf Morphology"](http://www.plantphysiol.org/content/170/3/1480.long) the shape of grapevine leaves had been compared between the 2013 and 2015 growing seasons. In the 2015 seasons, the distal sinus of the leaves was on average more pronounced than 2013. This is interesting because 2015 was the colder and drier of the two years, and in the paleorecord, plants from the colder, drier temperate regions (in contrast to the hotter, wetter tropics) have deeper lobes and serrations over millions of years of geologic time. This relationship between lobing and climate is so strong that fossil leaves used to be used to estimate ancient climates (before the use of radioisotopes).

For this midterm, you will be analyzing a third year of data from 2016. And for the final, you will be analyzing a fourth year of data from 2017 (but it still needs to be collected). The data consists of ***landmarks***, x and y coordinate values that correspond to the same points in every leaf. The landmark points are manually placed by hand. For example, for the 2016 data, 46,515 points for 2,215 grapevine leaves were placed by hand. The points can either represent the actual pixel corrdinates in the original scan, and therefore are useful for measuring size (especially for the allometry exercises for Group 2). Or, a Generalized Procrustes Analysis can be performed in which the points are superimposed on each other. During a Procrustes Analysis points are scaled, translated, rotated, flipped to superimpose points. This analysis is useful for analyzing shape from the x and y coordinate values directly, but it destroys size information.

The figure below shows the landmarks and features of the leaf that we will be studying.

The following datasets are available in the ***"Midterm"*** section of D2L:
    
-  Procrustes_tips.csv: The Procrustes adjusted landmark values for leaves matched by counting from the growing tip. These leaves are matched based on numbering ***from the tip of the growing vine***. Every leaf has a matching leaf from the ***same vine, from the same node (counting from tip) for each of three years.*** The leaves in this dataset are super-imposed, and the x and y coordinate values can be compared for each leaf against any other. Size information for the leaves has been lost. The columns are as follows: `tip_vine`: the node number counting from the tip and vine separated by 'x', `vine`: the vine number, `year`: the year of collection (2013, 2015, or 2016), `vine_year`: the vine and year seprated by 'x', `species`: the species, `from_tip`: the node number conunting from the tip, `from_base`: the node number counting from the based of the vine, `count`: the overall number of leaves for the vine, `base_vine`: the node number counting from the base and vine separated by 'x', `PCs 1-42`: 42 PC values for the Procrustes coordinates matched by tip count, `x and y, 1-21`: the Procrusted-adjusted coordinate values for landmarks 1-21.


- Procrustes_base.csv: Exactly like the tips dataset above, except ***leaves are matched by vine and node by counting from the base of the vine***.


- Allometry_tips.csv: In this dataset, the coordinate values are not Procrustes adjusted, they are the original pixel coordinate values from the scan. Because of this, the leaf shapes are not superimposed, but they do retain the original size information. The columns in this dataset are exactly the same as the Procrustes dataset, except there are no PCA values. Additionally, the column `px2_cm2` is a pixels^2 / cm^2 conversion that is necessary to calculate absolute area. ***The leaves in this dataset are matched by vine and node for each year by counting from the growing tip.***


- Allometry_base.csv: Exactly like the tips dataset above, except ***leaves are matched by vine and node by counting from the base of the vine***.


- climate_data.csv: For each growing season, the weather from March the previous year through June of the year of collection will be analyzed (for example, for 2013 data we would analyze March 2012 - June 2013). This is because the leaves for the next year are already present in the buds of the current year (they are just really tiny!) and so therefore we analyze two years worth of information to see the effects on leaf shape. This dataset provides daily values for average air temperature in Fahrenheit (`avg_air_temp_F`), max air temperature in F (`max_air_temp_F`), min air temperature in F (`min_air_temp_F`), total precipitation in inches (`total_precip_IN`), the number of hours in the day the leaf was wet (`leaf_wetness_hours`), hours for which humidity was greater than 90% (`rh_hrs_at_least_90_percent`), average wind speed in mph (`avg_wind_speed_mph`), and solar radiation in Langleys (`solar_rad_langleys`).

# Group 2: Allometry and climate data

1. Allometry is how shape changes relative to size. As shown in [Fig. 2A-C](http://www.plantphysiol.org/content/170/3/1480.long), if the natural log of a specific leaf area is plotted against natural log of overall leaf area, a linear relationship often results. If this relationship is preserved across a factor (like species or year) the relationship is called *isometric*. However, if a factor changes a slope, then it is allometric: the relationship between shape and size is different.
