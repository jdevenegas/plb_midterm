**1. Data wrangling and csv formatting** <br>
Several steps were undergone to format data for this pipeline and for each type of prediction. <br>
(a) observe given data, fix typos, and get rid of columns that aren't going to be used as features <br>
(b) create csv's for each prediction type, getting rid of irrelevant labels to that prediction type <br>
(c) replace integer labels with equivalent strings (for compatability with pipeline) <br>

Throughout, we observe what the smallest classes are in order to know how many instances each class has in our balanced training set.

**1A. Observe data, fix typos, and get rid of PC columns**

```python
import numpy as np
import pandas as pd
```

```python
# import df's
base = pd.read_csv('Procrustes_base.csv')
tips = pd.read_csv('Procrustes_tips.csv')
```

```python
# observe geneal structure of data
base.head()
```

```python
# Remove weird species names.
print(base.species.unique())

# There exist rows with species== 'Vitis_sp.' but we need to get rid of the period

base['species'] = np.where((base.species == 'Vitis_sp.'),'Vitis_sp',base.species)
tips['species'] = np.where((tips.species == 'Vitis_sp.'),'Vitis_sp',tips.species)

# There are some Vitis_x_novae-angliae and some Vitis_x_novae_angliae

base['species'] = np.where((base.species == 'Vitis_x_novae-angliae'),'Vitis_x_novae_angliae',base.species)
tips['species'] = np.where((tips.species == 'Vitis_x_novae-angliae'),'Vitis_x_novae_angliae',tips.species)
```

```python
# observe what columns are contained in dfs 
print('Procrustes_base.csv column names are: {}'.format(list(base.columns)))
```

```python
print('Procrustes_tips.csv column names are: {}'.format(list(tips.columns)))
```

```python
# get rid of columns with data irrelevant to ML 
for col in list(base.columns):
    if 'PC' in col:
        base.drop(labels=col,axis=1,inplace=True)
        
for col in list(tips.columns):
    if 'PC' in col:
        tips.drop(labels=col,axis=1,inplace=True)
```

```python
# check that columns were successfully eliminated
base.head()
```

```python
tips.head()
```

**1B. Create csv's for each prediction type**

```python
# drop labels and rename class column
# using base_vine or tip_vine as the unique identifier for all samples
# prediction of species
species = base.copy()
species.drop(labels=['vine','year','vine_year','from_tip','from_base','count','tip_vine'],axis=1,inplace=True)
species.rename(columns={'species':'Class'},inplace=True)

# prediction of year
year = base.copy()
year.drop(labels=['vine','vine_year','species','from_tip','from_base','count','tip_vine'],axis=1,inplace=True)
year.rename(columns={'year':'Class'},inplace=True)

# prediction of node from base 
nodeBase = base.copy()
nodeBase.drop(labels=['vine','year','vine_year','species','from_tip','count','tip_vine'],axis=1,inplace=True)
nodeBase.rename(columns={'from_base':'Class'},inplace=True)

# prediction of node from tip
nodeTip = tips.copy()
nodeTip.drop(labels=['vine','year','vine_year','species','from_base','count','base_vine'],axis=1,inplace=True)
nodeTip.rename(columns={'from_tip':'Class'},inplace=True)
```

```python
# check for duplicate ID's -- need to make sure each instance has a unique identifier
IDs = {}
for index,row in species.iterrows():
    if row['base_vine'] not in IDs:
        IDs[row['base_vine']] = 1
    else: IDs[row['base_vine']] += 1

#for key in IDs:
    #if IDs[key] != 1:
        #print(key)
# the above code is commented out because it would appear that every ID has 3 instances!

# base_vine and tip_vine are not unique identifiers because there is data from three years 
# use numerical index as instance identifiers 
# drop base_vine and tip_vine
species.drop(labels='base_vine',axis=1,inplace=True)
year.drop(labels='base_vine',axis=1,inplace=True)
nodeBase.drop(labels='base_vine',axis=1,inplace=True)
nodeTip.drop(labels='tip_vine',axis=1,inplace=True)
```

```python
# check what species we have 
speciesDict={}
for index,row in species.iterrows():
    if row['Class'] not in speciesDict:
        speciesDict[row['Class']]=1
    else: speciesDict[row['Class']] +=1
            
print(speciesDict)
```

```python
# drop hybrids
hybrids = ['Vitis_x_doaniana','Vitis_x_novae_angliae','Vitis_x_champinii','Vitis_x_andersonii']
outgroup = 'Ampelopsis_brevipedunculata'
species_noHybrids = species.copy()
for hybrid in hybrids:
    indices = species_noHybrids.index[species_noHybrids['Class'] == hybrid].tolist()
    species_noHybrids.drop(index=indices,inplace=True)
```

```python
# check that hybrids are gone
species_noHybridsDict={}
for index,row in species_noHybrids.iterrows():
    if row['Class'] not in species_noHybridsDict:
        species_noHybridsDict[row['Class']]=1
    else: species_noHybridsDict[row['Class']] +=1
            
print(species_noHybridsDict)
```

```python
# make another that excludes outgroup
outgroup = 'Ampelopsis_brevipedunculata'
species_noOutgroup = species_noHybrids.copy()
indices_noOutgroup = species_noOutgroup.index[species_noOutgroup['Class'] == outgroup].tolist()
species_noOutgroup.drop(index=indices_noOutgroup,inplace=True)

species_noOutgroupDict = {}
for index,row in species_noOutgroup.iterrows():
    if row['Class'] not in species_noOutgroupDict:
        species_noOutgroupDict[row['Class']]=1
    else: species_noOutgroupDict[row['Class']] +=1
            
print(species_noOutgroupDict)
```

```python
# write out the files
species.to_csv('species_prediction_dataset.csv',index=True,index_label='Instance')
species_noHybrids.to_csv('species_noHybrids_prediction_dataset.csv',index=True,index_label='Instance')
species_noOutgroup.to_csv('species_noOutgroupOrHybrids_prediction_dataset.csv',index=True,index_label='Instance')
year.to_csv('year_prediction_dataset.csv',index=True,index_label='Instance')
nodeBase.to_csv('nodes_from_base_prediction_dataset.csv',index=True,index_label='Instance')
nodeTip.to_csv('nodes_from_trip_prediction_dataset.csv',index=True,index_label='Instance')
```

```python
# make dataset with just rupestris and riparia
species = pd.read_csv('species_noOutgroupOrHybrids_prediction_dataset.csv')
species.head()
```

```python
# drop everything but rupestris and riparia
for index,row in species.iterrows():
    if row['Class'] != 'Vitis_rupestris' and row['Class'] != 'Vitis_riparia':
        species.drop(labels=index,axis=0,inplace=True)
species.head()
```

```python
# write out the file 
species.to_csv('species_RupRip_prediction.csv',index=False)
```

```python
# check what the smallest group is and how many are in it: nodes from tip
num_classes_tip = {}
with open('nodes_from_tip_strings_prediction_dataset.csv') as inputf:
    reader = csv.DictReader(inputf)
    
    for row in reader: 
        if row['Class'] not in num_classes_tip:
            num_classes_tip[row['Class']]=1
        else: num_classes_tip[row['Class']] +=1
            
print(num_classes_tip)
```

**Conclusions and closing notes** <br>
Performance metrics can be seen further in the output files ending in SVM_results.txt, and are also outlined in the powerpoint presentation. <br>
Overall, models for species and year perform quite well, whereas models for node prediction are not as successful. <br>
Note that, for multiclass models, the expectation for accuracy *and* for the F1 measure if the model were performing no better than random guessing, would be 1/(number of classes). For example, when predicting node, the expected F1 would be 1/13, or 0.0769. <br>
While the x-y coordinates perform quite well when differentiating species and year, it would be interesting to try these models with other types of features (perhaps persistent homology,or climate for the prediction of species or node) that may, when the importance of those features in prediction is determined, give us some biological insights to the mechanisms of leaf morhpology differences.
