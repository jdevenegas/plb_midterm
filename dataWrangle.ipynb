Wrangling input data for use in ML Pipeline

```python
import pandas as pd
```

```python
# import df's
base = pd.read_csv('Procrustes_base.csv')
tips = pd.read_csv('Procrustes_tips.csv')
```

```python
# observe geneal structure of data
base.head()
```

```python
tips.head()
```

```python
# observe what columns are contained in df 
print('Procrustes_base.csv column names are: {}'.format(list(base.columns)))
```

```python
print('Procrustes_tips.csv column names are: {}'.format(list(tips.columns)))
```

```python
# get rid of columns with data irrelevant to ML 
for col in list(base.columns):
    if 'PC' in col:
        base.drop(labels=col,axis=1,inplace=True)
        
for col in list(tips.columns):
    if 'PC' in col:
        tips.drop(labels=col,axis=1,inplace=True)
```

```python
base.head()
```

```python
tips.head()
```

```python
# drop labels and rename class column
# using base_vine or tip_vine as the unique identifier for all samples
# prediction of species
species = base.copy()
species.drop(labels=['vine','year','vine_year','from_tip','from_base','count','tip_vine'],axis=1,inplace=True)
species.rename(columns={'species':'Class'},inplace=True)

# prediction of year
year = base.copy()
year.drop(labels=['vine','vine_year','species','from_tip','from_base','count','tip_vine'],axis=1,inplace=True)
year.rename(columns={'year':'Class'},inplace=True)

# prediction of node from base 
nodeBase = base.copy()
nodeBase.drop(labels=['vine','year','vine_year','species','from_tip','count','tip_vine'],axis=1,inplace=True)
nodeBase.rename(columns={'from_base':'Class'},inplace=True)

# prediction of node from tip
nodeTip = tips.copy()
nodeTip.drop(labels=['vine','year','vine_year','species','from_base','count','base_vine'],axis=1,inplace=True)
nodeTip.rename(columns={'from_tip':'Class'},inplace=True)
```

```python
# write out the files
species.to_csv('species_prediction_dataset.csv',index=False)
year.to_csv('year_prediction_dataset.csv',index=False)
nodeBase.to_csv('nodes_from_base_prediction_dataset.csv',index=False)
nodeTip.to_csv('nodes_from_trip_prediction_dataset.csv',index=False)
```
